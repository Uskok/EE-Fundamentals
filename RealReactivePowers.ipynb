#Instantaneous Power $p(t)$ and Real Power $P$

Let the instantaneous voltage and current be
$$\begin{align}
v(t) &= V_A\cos(\omega t + \theta_v)\\
i(t) &= I_A\cos(\omega t + \theta_i)\\
\end{align}\tag{1}$$

The instantanous power is
$$p(t) = v(t) i(t) = V_AI_A\cos(\omega t + \theta_v)\cos(\omega t + \theta_i)\tag{2}$$

Applying the trig identity $\cos x\cos y = \frac{\cos(x+y)+\cos(x_y)}{2}$ to (2) yields
$$p(t) = \frac{V_AI_A}{2}\left[\cos(2\omega t+\theta_v+\theta_i) + \cos(\theta_v-\theta_i)\right]\tag{3}$$

From (3) it is easy to identify the constant term, which is commonly known as the real power or active power, denoted by $P$
$$P = \frac{V_AI_A}{2}\cos(\theta_v-\theta_i)\tag{4}$$

It is worth noting that, the reactive power, or the imaginary power, denoted by $Q$ is not so easy to identify from (3).

#Real Power $P$ and Phasors

With reference to (1), the corresponding votlage and current phasors are
$$\begin{align}
\hat V &= V_A e^{j\theta_v}\\
\hat I &= I_A e^{j\theta_i}
\end{align}\tag{5}$$

Let the voltage phasor $\hat V$ multiplied by the current phasor conjugate $\hat I^*$
$$\begin{align}
\hat V \hat I^* &= V_A e^{j\theta_v} I_A e^{-j\theta_i}\\
 &= V_A I_A e^{j(\theta_v-\theta_i)}\\
 &= V_A I_A \cos(\theta_v-\theta_i) + jV_A I_A \sin(\theta_v-\theta_i)
\end{align}\tag{6}$$

By comparison of (4) and (6), the active power $P$ can be determined by phasors as
$$P = \frac{1}{2}Re(\hat V \hat I^*)\tag{7}$$
The factor $1\over2$ can be emlinated by using rms values as as the phasor magnitudes. Let's keep the factor in this discussion.


#Split Current $\hat I$ into Two Components $\hat I_P$ and $\hat I_Q$

Let the current phasor be rewritten as
$$\begin{align}
\hat I &= I_A e^{j[\theta_v - (\theta_v - \theta_i)]}\\
       &= \underbrace{I_A\cos(\theta_v - \theta_i) e^{j\theta_v}}_{\hat I_P}  \underbrace{-jI_A\sin(\theta_v - \theta_i) e^{j\theta_v}}_{\hat I_Q}\\
       &= \hat I_P + \hat I_Q
\end{align}\tag{8}$$

Let's redo (6) using splitted currents in (8), we have
$$\begin{align}
\hat V\hat I_P^* &= V_AI_A\cos(\theta_v - \theta_i)\\
\hat V\hat I_Q^* &= jV_AI_A\sin(\theta_v - \theta_i)
\end{align}\tag{9}$$


The current in time domain can be split in a similar fashion
$$\begin{align}
i_P(t) &= I_A\cos(\theta_v - \theta_i)\cos(\omega t+\theta_v)\\
i_Q(t) &= I_A\sin(\theta_v - \theta_i)\cos(\omega t+\theta_v-90^\circ)=I_A\sin(\theta_v - \theta_i)\sin(\omega t+\theta_v)
\end{align}\tag{10}$$

Now, the instantaneous power $p(t)$ is
$$\begin{align}
p(t) &= v(t) [i_P(t) + i_Q(t)]\\
     &= V_AI_A\cos(\theta_i - \theta_v)\cos^2(\omega t+\theta_v) + V_AI_A\sin(\theta_v - \theta_i)\cos(\omega t+\theta_v)\sin(\omega t+\theta_v)\\
     &= \underbrace{\frac{V_AI_A}{2}\cos(\theta_v - \theta_i)}_{P=\frac{\hat V\hat I_P^*}{2}} [\cos(2\omega t+2\theta_v)+1] + \underbrace{\frac{V_AI_A}{2}\sin(\theta_v - \theta_i)}_{Q = Im\left(\frac{\hat V\hat I_Q^*}{2}\right)}\sin(2\omega t+2\theta_v)
\end{align}\tag{11}$$
The amplitude of the second term in (11) is identified as the reactive power $Q$.

The physical interpretation of this current split can be the $\hat I_P$ for the current through a resistance while $\hat I_Q$ for the current through an inductance or capacitance. The resistance is in parallel with the inductance or the capacitance.

#Split Current $\hat V$ into Two Components $\hat V_P$ and $\hat V_Q$

Alternavetively, the voltage can split into two components, one in phase with the current and the other one in quadrature to the current.

Let the voltage phasor be rewritten as
$$\begin{align}
\hat V &= V_A e^{j[(\theta_v - \theta_i) + \theta_i]}\\
       &= \underbrace{V_A\cos(\theta_v - \theta_i) e^{j\theta_i}}_{\hat V_P} + \underbrace{jV_A\sin(\theta_v - \theta_i) e^{j\theta_i}}_{\hat V_Q}\\
       &= \hat V_P + \hat V_Q
\end{align}\tag{12}$$

Let's redo (6) using splitted currents in (12), we have
$$\begin{align}
\hat V_P\hat I^* &= V_AI_A\cos(\theta_v - \theta_i)\\
\hat V_Q\hat I^* &= jV_AI_A\sin(\theta_v - \theta_i)
\end{align}\tag{13}$$


The voltage in time domain can be split in a similar fashion
$$\begin{align}
v_P(t) &= V_A\cos(\theta_v - \theta_i)\cos(\omega t+\theta_i)\\
v_Q(t) &= V_A\sin(\theta_v - \theta_i)\cos(\omega t+\theta_i+90^\circ)=-I_A\sin(\theta_v - \theta_i)\sin(\omega t+\theta_i)
\end{align}\tag{14}$$

Now, the instantaneous power $p(t)$ is
$$\begin{align}
p(t) &=  [v_P(t) + v_Q(t)]i(t)\\
     &= V_AI_A\cos(\theta_v - \theta_i)\cos^2(\omega t+\theta_i) - V_AI_A\sin(\theta_i - \theta_v)\sin(\omega t+\theta_i)\cos(\omega t+\theta_i)\\
     &= \underbrace{\frac{V_AI_A}{2}\cos(\theta_v - \theta_i)}_{P=\frac{\hat V_P\hat I^*}{2}} [\cos(2\omega t+2\theta_i)+1] - \underbrace{\frac{V_AI_A}{2}\sin(\theta_v - \theta_i)}_{Q = Im\left(\frac{\hat V_Q\hat I^*}{2}\right)}\sin(2\omega t+2\theta_i)
\end{align}\tag{15}$$
The amplitude of the second term in (15) is identified as the reactive power $Q$.

The physical interpretation of this current split can be the $\hat V_P$ for the votlage across a resistance while $\hat V_Q$ for the voltage across an inductance or capacitance. The resistance is in sereies with the inductance or the capacitance.

```
import matplotlib
from os import linesep
from numpy import linspace,cos,pi,sin,random,exp,real, imag,heaviside,fft,array,conj
import matplotlib.pyplot as plt
from scipy.integrate import quad
from matplotlib import animation,rc
rc('animation', html='jshtml')
plt.rcParams.update({"font.family" : "sans serif","mathtext.fontset" : "cm"})

# draw dotted line between three sets of points in three different plots
def l2Ax(fig,ax1,ax2,ax3,ax4,p1,p2,p3,colors):
  transFigure = fig.transFigure.inverted()
  ls=[]
  for k in range(len(p1)):
    p1_g = transFigure.transform(ax1.transData.transform(p1[k]))
    p2_g = transFigure.transform(ax2.transData.transform(p2[k]))
    p3_g = transFigure.transform(ax3.transData.transform(p3[k]))
    if k<4:
      line1 = matplotlib.lines.Line2D((p1_g[0],p2_g[0]),(p1_g[1],p2_g[1]),
                          transform=fig.transFigure,ls='--',color=colors[k], lw=1)
      line2 = matplotlib.lines.Line2D((p2_g[0],p3_g[0]),(p2_g[1],p3_g[1]),
                          transform=fig.transFigure,ls='--',color=colors[k], lw=1)
      ls.extend([line1,line2])
    else:
      p3_g = transFigure.transform(ax4.transData.transform(p3[k]))
      line1 = matplotlib.lines.Line2D((p1_g[0],p3_g[0]),(p1_g[1],p3_g[1]),
                          transform=fig.transFigure,ls='--',color=colors[k], lw=1)
      ls.append(line1)
  fig.lines= ls
  return

def drawMirror(ax,colors):
  ax.plot([-1.5,1.5], [-1.5,1.5], color=colors[-2], lw=3)
  for i in range(100):
    ax.plot([-1.6+i*3/100, -1.53+i*3/100],[-1.5+i*3/100,-1.5+i*3/100], color='lightgray', lw=1)
  return

# def fixedVectorsDraw(ax,vecs,txt,colors): #draw fixed phasor
#   ax.clear()
#   ax.set_xlim(-1.5,1.5)
#   ax.set_ylim(-1.5,1.5)
#   ax.axis('off')
#   for k in range(len(vecs)):
#     ax.arrow(0,0,vecs[k].real,vecs[k].imag,
#              width = 0.01,head_width=0.04,length_includes_head=True,color=colors[k])
#     ax.text(vecs[k].real,vecs[k].imag,txt[k],size=24,color=colors[k])
#   if len(vecs)>3:
#     ax.plot([vecs[1].real,vecs[2].real],[vecs[1].imag,vecs[2].imag],'--',color='gray',lw=1)
#     ax.plot([vecs[1].real,vecs[3].real],[vecs[1].imag,vecs[3].imag],'--',color='gray',lw=1)
#   return

def dynamicVectorsDraw(ax,vecs,t,txts,colors): #draw rotating phasor
  ax.clear()
  ax.set_xlim(-1.5,1.5)
  ax.set_ylim(-1.5,1.5)
  ax.axis('off')
  for k in range(len(vecs)):
    ax.arrow(0,0,vecs[k].real,vecs[k].imag,width = 0.01,head_width=0.04,
            length_includes_head=True,color=colors[k])
    ax.text(vecs[k].real,vecs[k].imag,txts[k],size=24,color=colors[k])
    vec_traj = vecs[k]*exp(1j*2*pi*t)#trajectory of the vector rotation
    if len(vecs)<=3 and k<2:
      vec_r = vec_traj[-1] #rotating vector
      ax.arrow(0,0,vec_r.real,vec_r.imag,width = 0.01,head_width=0.04,
            length_includes_head=True,color=colors[k],alpha=0.6)
      ax.text(vec_r.real,vec_r.imag,txts[k]+'$e^{j\omega t}$',size=20,color=colors[k],alpha=.8)
      ax.plot(vec_traj.real,vec_traj.imag,color=colors[k],lw=2,alpha=.3)
    elif len(vecs)>3 and k<3:
      vec_r = vec_traj[-1] #rotating vector
      ax.arrow(0,0,vec_r.real,vec_r.imag,width = 0.01,head_width=0.04,
            length_includes_head=True,color=colors[k],alpha=0.6)
      ax.text(vec_r.real,vec_r.imag,txts[k]+'$e^{j\omega t}$',size=20,color=colors[k],alpha=.8)
      ax.plot(vec_traj.real,vec_traj.imag,color=colors[k],lw=2,alpha=.3)
  if len(vecs)>5:
    ax.plot([vecs[5].real,vecs[2].real],[vecs[5].imag,vecs[2].imag],'--',color='gray',lw=1)
    ax.plot([vecs[1].real,vecs[5].real],[vecs[1].imag,vecs[5].imag],'--',color='gray',lw=1)
  return


def texts(ax,xy,offsets,txts,colors):
  ax.clear()
  ax.set_xlim(-.1,1.1)
  ax.set_ylim(-1.5,1.5)
  ax.axis('off')
  for k in range(len(txts)):
    ax.text(xy[0]+offsets[k],xy[1],txts[k],size=18,color=colors[k])
  return
#parameters
# clst=['lightskyblue','violet','aqua','cornflowerblue','yellow','w',(.1,.1,.1)] #colors dark bg
clst=['mediumblue','m','brown','c','r',(.1,.1,.1),'w'] #colors light bg
Nf = 300
t=linspace(0,1,Nf)
ewt = exp(1j*2*pi*t)
#voltage and current phasors
V_A = 1
theta_v = pi/7
Vhat  = V_A*exp(1j*theta_v)# voltage phasor
I_A = 1.2
theta_i = -pi/6
Ihat =I_A*exp(1j*theta_i) #current phasor
#split of voltage and current phasors
VPhat = V_A*cos(theta_v-theta_i)*exp(1j*theta_i)
VQhat = V_A*sin(theta_v-theta_i)*exp(1j*(theta_i+pi/2))
IPhat = I_A*cos(theta_i-theta_v)*exp(1j*theta_v)
IQhat = I_A*sin(theta_i-theta_v)*exp(1j*(theta_v+pi/2))

#time-domain waveforms
vt = (Vhat*ewt).real #v(t)
vPt = (VPhat*ewt).real #v_P(t)
vQt = (VQhat*ewt).real #v_Q(t)
it = (Ihat*ewt).real #i(t)
iPt = (IPhat*ewt).real #i_P(t)
iQt = (IQhat*ewt).real #i_Q(t)
vt1 = (Vhat*ewt).imag
vPt1 = (VPhat*ewt).imag
vQt1 = (VQhat*ewt).imag
it1 = (Ihat*ewt).imag
iPt1 = (IPhat*ewt).imag
iQt1 = (IQhat*ewt).imag
pt =vt*it #p(t)
pt_vPi = vPt*it
pt_vQi = vQt*it
pt_viP = vt*iPt
pt_viQ = vt*iQt

fig = plt.figure(figsize=(16,9),facecolor=clst[-1])
fig.tight_layout()
ax_frame = [[[0, 0, .28125, .5], [-1.5,1.5], [-1.5,1.5]], #complex plane
            [[0, 0, .28125, .5], [-1.5,1.5], [-1.5,1.5]], #animated vectors
            [[0, .5, .28125, .5], [-1.5,1.5], [-1.5,1.5]], #mirror
            [[.3, .5, .7, .5], [-.1,1.1], [-1.5,1.5]], #upper right for texts
            [[.3, 0, .7, .5], [-.1,1.1], [-1.5,1.5]], #lower right for texts
            [[.3, .5, .7, .5], [-.1,1.1], [-1.5,1.5]], #v(t),i_P(t),i_Q(t),v(t)i_P(t)
            [[.3, 0, .7, .5], [-.1,1.1], [-1.5,1.5]], #v(t)i_Q(t)
            ]# [pos-boundary, xlim, ylim] for subplots 16:9
# ax_frame = [[[0, 0, .5, .5], [-1.5,1.5], [-1.5,1.5]], #complex plane
#             [[0, 0, .5, .5], [-1.5,1.5], [-1.5,1.5]], #animated vectors
#             [[0, .5, .5, .5], [-1.5,1.5], [-1.5,1.5]], #mirror
#             [[.5, .5, .5, .5], [-.1,1.1], [-1.5,1.5]], #upper right for texts
#             [[.5, 0, .5, .5], [-.1,1.1], [-1.5,1.5]], #lower right for texts
#             [[.5, .5, .5, .5], [-.1,1.1], [-1.5,1.5]], #v(t),i_P(t),i_Q(t),v(t)i_P(t)
#             [[.5, 0, .5, .5], [-.1,1.1], [-1.5,1.5]], #v(t)i_Q(t)
#             ]# [pos-boundary, xlim, ylim] for subplots 9:9
ax_lbl=[['$Im$', '$Re$'],
        ['', ''],
        ['', ''],
        ['', ''],
        ['', ''],
        ['', '$t$'], # time domain plot
        ['', '$t$'],
        ] #variables for yx axes of each subplot

ax_lst=[] #axis list or the list of subplots
ytxt_lst=[] #list of y texts
line_lst = [] #array containing all the line segments
for k in range(len(ax_frame)):
  xn,xm,yn,ym = ax_frame[k][1]+ax_frame[k][2]
  ax=fig.add_axes(ax_frame[k][0],xlim=[xn,xm], ylim=[yn,ym],fc='none') #no fc
  ax.axis('off') #turn off axis frames
  ax_lst.append(ax)
  if k!=2:
    ax.annotate("", (xm, 0), (xn, 0),arrowprops={'arrowstyle':"->",'color':clst[-2]}) #x-axis
    ax.annotate("", (0,ym), (0,yn),arrowprops={'arrowstyle':"->",'color':clst[-2]}) #y-axis
  if k>4:
    for i in range(5):
      line_lst.append(ax.plot([], [],color=clst[i], lw=2)[0])

  # line_lst.append(ax.plot([], [],color=clst[k], lw=2)[0])
  ax.text(xm-.02,0,ax_lbl[k][-1],size=20,va='top',ha='right',color=clst[-2]) #x-axis label
  if k==0:
    ytxt_lst.append(ax.text(0.01,ym,ax_lbl[k][0],color=clst[-2],size=20,va='center')) #y label
  # else:
  #   ytxt_lst.append() #y label

# animation function. This is called sequentially
def animate(i):
  ii = i%Nf
  k = int(i/Nf)
  if ii==0: #for monitoring
    print(k)
  if k ==0:
    if ii==0:
      texts(ax_lst[3],[0,ax_frame[3][2][1]-.2],[0,0.08,0.08*2],['$v(t)$,','$i(t)$'],clst)
    dynamicVectorsDraw(ax_lst[1],[Vhat,Ihat],t[:ii+1],['$\hat V$','$\hat I$'],clst)
    l2Ax(fig,ax_lst[0],ax_lst[2],ax_lst[5],ax_lst[6],
     [[vt[ii],vt1[ii]],[it[ii],it1[ii]]],
     [[vt[ii],vt[ii]],[it[ii],it[ii]]],
     [[t[ii],vt[ii]],[t[ii],it[ii]]],clst)
    line_lst[0].set_data(t[:ii+1],vt[:ii+1])
    line_lst[1].set_data(t[:ii+1],it[:ii+1])
  elif k ==1:
    if ii==0:
      texts(ax_lst[3],[0,ax_frame[3][2][1]-.2],[0,0.08,0.08*2],['$v(t)$,','$i(t)$,','$v(t)i(t)$'],clst)
      ax_lst[4].text(.55,0.9, r'It is difficult to identify the reactive power component from $v(t)i(t)$.',size=16,ha='center',color=clst[5])
    dynamicVectorsDraw(ax_lst[1],[Vhat,Ihat,Vhat*conj(Ihat)*.5],t[:ii+1],['$\hat V$','$\hat I$',r'$\frac{\hat V\hat I^*}{2}$'],clst)
    l2Ax(fig,ax_lst[0],ax_lst[2],ax_lst[5],ax_lst[6],
     [[vt[ii],vt1[ii]],[it[ii],it1[ii]],[(.5*Vhat*conj(Ihat)).real,(.5*Vhat*conj(Ihat)).imag]],
     [[vt[ii],vt[ii]],[it[ii],it[ii]],[(.5*Vhat*conj(Ihat)).real,(.5*Vhat*conj(Ihat)).real]],
     [[t[ii],vt[ii]],[t[ii],it[ii]],[t[ii],(.5*Vhat*conj(Ihat)).real]], clst)
    line_lst[0].set_data(t[:ii+1],vt[:ii+1])
    line_lst[1].set_data(t[:ii+1],it[:ii+1])
    line_lst[2].set_data(t[:ii+1],pt[:ii+1])
  elif k ==2:
    if ii==0:
      texts(ax_lst[3],[0,ax_frame[3][2][1]-.2],[0,0.08,0.08*2,0.08*3,0.08*3],['$v(t)$,','$i_P(t)$,','$i_Q(t)$,','$v(t)i_P(t)$'],clst)
      texts(ax_lst[4],[0,ax_frame[4][2][1]-.2],[0],['$v(t)i_Q(t)$'],clst[4])
      ax_lst[4].text(.55,0.9, r'The amplitude of $v(t)i_Q(t)$ represents the reactive power $Im\left(\frac{\hat V\hat I_Q^*}{2}\right)$=$Im\left(\frac{\hat V\hat I^*}{2}\right)$',size=16,ha='center',color=clst[5])
    dynamicVectorsDraw(ax_lst[1],
     [Vhat,IPhat,IQhat,Vhat*conj(IPhat)*.5,Vhat*conj(IQhat)*.5,Ihat],t[:ii+1],
     ['$\hat V$','$\hat I_P$','$\hat I_Q$',r'$\frac{\hat V\hat I_P^*}{2}$',r'$\frac{\hat V\hat I_Q^*}{2}$','$\hat I$'],clst)
    l2Ax(fig,ax_lst[0],ax_lst[2],ax_lst[5],ax_lst[6],
     [[vt[ii],vt1[ii]],[iPt[ii],iPt1[ii]],[iQt[ii],iQt1[ii]],[(.5*Vhat*conj(IPhat)).real,0],[0,(.5*Vhat*conj(IQhat)).imag]],
     [[vt[ii],vt[ii]],[iPt[ii],iPt[ii]],[iQt[ii],iQt[ii]],[(.5*Vhat*conj(IPhat)).real,(.5*Vhat*conj(IPhat)).real],[0,(.5*Vhat*conj(IQhat)).imag]],
     [[t[ii],vt[ii]],[t[ii],iPt[ii]],[t[ii],iQt[ii]],[t[ii],(.5*Vhat*conj(IPhat)).real],[t[ii],(.5*Vhat*conj(IQhat)).imag]], clst)
    line_lst[0].set_data(t[:ii+1],vt[:ii+1])
    line_lst[1].set_data(t[:ii+1],iPt[:ii+1])
    line_lst[2].set_data(t[:ii+1],iQt[:ii+1])
    line_lst[3].set_data(t[:ii+1],pt_viP[:ii+1])
    line_lst[9].set_data(t[:ii+1],pt_viQ[:ii+1])
  elif k ==3:
    if ii==0:
      texts(ax_lst[3],[0,ax_frame[3][2][1]-.2],[0,0.08,0.08*2,0.08*3,0.08*3],['$i(t)$,','$v_P(t)$,','$v_Q(t)$,','$i(t)v_P(t)$'],clst)
      texts(ax_lst[4],[0,ax_frame[4][2][1]-.2],[0], ['$i(t)v_Q(t)$'],clst[4])
      ax_lst[4].text(.55,0.9, r'The amplitude of $i(t)v_Q(t)$ represents the reactive power $Im\left(\frac{\hat V_Q\hat I^*}{2}\right)$=$Im\left(\frac{\hat V\hat I^*}{2}\right)$',size=16,ha='center',color=clst[5])
    dynamicVectorsDraw(ax_lst[1],
     [Ihat,VPhat,VQhat,VPhat*conj(Ihat)*.5,VQhat*conj(Ihat)*.5,Vhat],t[:ii+1],
     ['$\hat I$','$\hat V_P$','$\hat V_Q$',r'$\frac{\hat V_P\hat I^*}{2}$',r'$\frac{\hat V_Q\hat I^*}{2}$','$\hat V$'],clst)
    l2Ax(fig,ax_lst[0],ax_lst[2],ax_lst[5],ax_lst[6],
     [[it[ii],it1[ii]],[vPt[ii],vPt1[ii]],[vQt[ii],vQt1[ii]],[(.5*VPhat*conj(Ihat)).real,0],[0,(.5*VQhat*conj(Ihat)).imag]],
     [[it[ii],it[ii]],[vPt[ii],vPt[ii]],[vQt[ii],vQt[ii]],[(.5*VPhat*conj(Ihat)).real,(.5*VPhat*conj(Ihat)).real],[0,(.5*VQhat*conj(Ihat)).imag]],
     [[t[ii],it[ii]],[t[ii],vPt[ii]],[t[ii],vQt[ii]],[t[ii],(.5*VPhat*conj(Ihat)).real],[t[ii],(.5*VQhat*conj(Ihat)).imag]], clst)
    line_lst[0].set_data(t[:ii+1],it[:ii+1])
    line_lst[1].set_data(t[:ii+1],vPt[:ii+1])
    line_lst[2].set_data(t[:ii+1],vQt[:ii+1])
    line_lst[3].set_data(t[:ii+1],pt_vPi[:ii+1])
    line_lst[9].set_data(t[:ii+1],pt_vQi[:ii+1])


  return
drawMirror(ax_lst[2],clst)
anim = animation.FuncAnimation(fig, animate, frames=Nf*4, interval=20)
# anim #uncomment to generate animation in the output area
# to save the animation, uncomment the following three lines
fn = "RealReactivePowers.mp4"
writervideo = animation.FFMpegWriter(fps=25)
anim.save(fn, writer=writervideo,dpi = 120)

#debug
# print(len(trajrec[0]))
```
